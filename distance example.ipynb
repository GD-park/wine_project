```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

 
df = pd.read_csv('final_preprocessing.csv')


data_pool = np.array(df.ix[:,1:])
x_new = data_pool[2]

distance = np.linalg.norm(data_pool - x_new, axis=1)
df['distance_2']=distance

cosine = cosine_similarity(data_pool, x_new)
df['cosine_2']=cosine

 
df.to_csv('result_add_cosinedistance.csv')
```
