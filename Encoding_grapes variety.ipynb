```python
import numpy as np
import pandas as pd
import re
from collections import OrderedDict
```

```python
tn = pd.read_csv('wine_data_170302.csv')
tn.ix[:,["id","name"]]
```

```python
sm = pd.read_csv('varieties0313.csv')
encd_df = pd.DataFrame(sm.columns)

for k in range(100):
    encd = np.zeros(20)                    #각 와인별
    for j in range(20):                    #각 카테고리별
        for i in sm.ix[:,j]:               #카테고리의 품종별
            if str(i) in tn.ix[k,1]:       
                encd[j] += 1
    encd_df[tn.ix[k,0]] = encd
encd_df.to_csv('encoding0223.csv')
```
